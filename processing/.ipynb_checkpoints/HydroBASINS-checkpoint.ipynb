```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import netCDF4
import geopandas
from matplotlib.colors import ListedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable

#%matplotlib notebook

lim_area=6.5e4
```

# Level 3 basins from HydroATLAS

Download level 3 basins from HydroSHEDS geometry completed with multiple variables [HydroATLAS](https://hydrosheds.org/page/hydroatlas).

```python
level=3

basins_atlas=geopandas.read_file("../datasets/basins/BasinATLAS_v10_lev0{}.shp".format(level))

# remove Greenland and select only basins larger than 65 000 km2
basins3=basins_atlas.loc[(basins_atlas['HYBAS_ID']<9000000000)&(basins_atlas['SUB_AREA']>=lim_area)]

# remove Arctic islands
basins3=basins3.loc[basins3['HYBAS_ID']<=8030016860]

print('We have {} level 3 basins larger than {} km^2'.format(basins3.shape[0],lim_area))

basins3.head()
```

Colors of each climate zone following the documentation

```python
climate_colors=["#beffe7","#7bf5c9","#38a702","#7b8cf3","#004da6","#0085a8","#00724c",
             "#99e600","#d0ff72","#aa66cd","#cd6598","#ff7f7e",
             "#f579b6","#d69dbd","#ffa77f","#ffeabd","#ffffbf","#ffbee8"]
climate_names=["Arctic 1","Arctic 2","Cold and mesic","Cold and wet",
               "Extremely cold and wet 1","Extremely cold and wet 2","Extremely cold and mesic",
               "Cool temperate and dry","Cool temperate and xeric","Cool temperate and moist",
               "Warm temperate and mesic","Warm temperate and xeric",
               "Hot and mesic","Hot and dry","Hot and arid",
               "Extremely hot and arid","Extremely hot and xeric","Extremely hot and moist"]
```

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.set_aspect('equal')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
basins3.plot(ax=ax,edgecolor='white',linewidth=0.3,column='clz_cl_smj',
             vmin=1,vmax=18,cmap=ListedColormap(climate_colors),
                  legend=True,cax=cax)
ax.set_facecolor("grey")
ax.set_title('{} level 3 basins larger than {:.0f} $km^2$'.format(basins3.shape[0],lim_area))
plt.show()
```

Explode basins and dissolve to remove tiny islands

```python
expl_basins3=basins3.explode()
expl_basins3['AREA']=expl_basins3['geometry'].to_crs('epsg:6933').area/1e6
```

```python
large_expl_basins3=expl_basins3.loc[expl_basins3['AREA']>=1000]
new_basins3=large_expl_basins3.dissolve(by='HYBAS_ID')
```

```python
new_basins3['AREA']=new_basins3['geometry'].to_crs('epsg:6933').area/1e6
```

```python
# check that we did not lose basins while removing islands
new_basins3.loc[new_basins3['AREA']>=lim_area].shape
```

Add variables that may explain clustering

```python
save_basins3=new_basins3[['AREA','clz_cl_smj','lka_pc_sse','dor_pc_pva','geometry']].copy()
save_basins3['lka_pc_sse']=save_basins3['lka_pc_sse']/10
save_basins3['dor_pc_pva']=save_basins3['dor_pc_pva']/10
save_basins3.columns=['AREA','CLIMATE_DETAILS','PERCENT_LAKE','DEGREE_OF_REGULATION','geometry']
save_basins3
```

Add climate names

```python
climate_df=pd.DataFrame({'CLIMATE_DETAILS':np.arange(1,19),
                        'CLIMATE_NAME':climate_names})
climate_df.set_index('CLIMATE_DETAILS',inplace=True)
save_basins3=save_basins3.join(climate_df,on='CLIMATE_DETAILS')
save_basins3
```

Plot basins with simplified boundaries

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.set_aspect('equal')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
save_basins3.plot(ax=ax,edgecolor='white',linewidth=0.5,column='CLIMATE_DETAILS',
                 vmin=np.min(save_basins3['CLIMATE_DETAILS']),vmax=18,
                  cmap=ListedColormap(climate_colors[np.min(save_basins3['CLIMATE_DETAILS'])-1:]),
                  legend=True,cax=cax)
ax.set_facecolor("grey")
ax.set_title('{} simplified level 3 basins larger than {:.0f} $km^2$'.format(new_basins3.shape[0],lim_area))
#ax.set_xlim([96.5,106])
#ax.set_ylim([0,17.5])
plt.show()
```

### Simplified climate

```python
save_basins3['SIMPL_CLIMATE']=save_basins3['CLIMATE_DETAILS']

# cold and wet
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==5,'SIMPL_CLIMATE']=4
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==6,'SIMPL_CLIMATE']=4
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==7,'SIMPL_CLIMATE']=4

# temperate and dry
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==9,'SIMPL_CLIMATE']=8
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==12,'SIMPL_CLIMATE']=8

# temperate and moist
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==11,'SIMPL_CLIMATE']=10

# hot and dry
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==14,'SIMPL_CLIMATE']=16
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==15,'SIMPL_CLIMATE']=16
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==17,'SIMPL_CLIMATE']=16

# hot and wet
save_basins3.loc[save_basins3['SIMPL_CLIMATE']==18,'SIMPL_CLIMATE']=13
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.set_aspect('equal')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
save_basins3.plot(ax=ax,edgecolor='white',linewidth=0.2,column='SIMPL_CLIMATE',
                 vmin=1,vmax=18,
                  cmap=ListedColormap(climate_colors),
                  legend=True,cax=cax)
#ax.set_facecolor("grey")
ax.set_title('{} simplified level 3 basins larger than {:.0f} $km^2$'.format(new_basins3.shape[0],lim_area))
#ax.set_xlim([96.5,106])
#ax.set_ylim([0,17.5])
#fig.tight_layout()
plt.show()
```
